# Session 13: Data Wrangling and Visualization

Today's session is a quick tour through some of the data visualization methods for exploring datasets using Pandas and the built-in charting capabilities of Matplotlib.

We won't go into details of Matplotlib, but you can get a lot more detailed documentation on the Matplotlib website: http://matplotlib.org/

Pandas adds a layer on top of Matplotlib that makes it a bit easier to use for many common data visualization tasks, particularly ones that make it easier to interactively explore your data.

Documentation includes Chapter 8 of the Python for Data Analysis book, and the online documentation, available at: 
http://pandas.pydata.org/pandas-docs/stable/visualization.html#plotting-directly-with-matplotlib

API documentation is an excellent way to learn about the details of the arguments used for each of these methods: 
* http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.hist.html

* http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.plot.html

* http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.boxplot.html

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
rent_bay_filtered = pd.read_csv('data/rent_bay_filtered.csv').dropna()
```

```python
rent_bay_filtered['price'].hist(bins=20)
```

```python
rent_bay_filtered['price'].hist(bins=20, normed=True)
```

```python
rent_bay_filtered[rent_bay_filtered['bedrooms']<5].hist(column='price', by='bedrooms', figsize=(12,12), normed=True)
```

```python
bedcounts = rent_bay_filtered.groupby('bedrooms')['bedrooms'].count()
bedcounts.plot(kind='bar')
```

```python
bedcounts.plot(kind='barh')
```

```python
rent_bay_filtered['price'].plot(kind='kde')
```

```python
rent_bay_filtered['price'].plot(kind='kde', figsize=(10,10))
rent_bay_filtered['price'].hist(bins=100, alpha=0.3, color='k', normed=True)
```

```python
rent_bay_filtered['bedrooms'].hist()
```

```python
plt.scatter(rent_bay_filtered['bedrooms'], rent_bay_filtered['price'])
```

```python
rent_bay_filtered.plot(kind='scatter', x='bedrooms', y='price')
```

```python
rent_bay_filtered.plot(kind='hexbin', x='sqft', y='price', gridsize=100, xlim=(0,2000), ylim=(1000,4000))
```

```python
rent=rent_bay_filtered[['price','bedrooms','sqft']]
pd.scatter_matrix(rent, diagonal='kde', figsize=(12, 12), color='g', alpha=0.3)
```

```python
rent.boxplot(column='price', by='bedrooms', figsize=(12,12))
```
